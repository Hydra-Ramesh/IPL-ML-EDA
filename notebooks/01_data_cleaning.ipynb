```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../data/raw/ipl_matches.csv', low_memory=False)
df.shape
df.head()
```

```python
df=df.drop(columns=['Unnamed: 0'],errors='ignore')
```

```python
df
```

```python
df['date']=pd.to_datetime(df['date'], errors='coerce')
```

```python
df
```

```python
df=df.drop(columns=['power_surge_start'], errors='ignore')
```

```python
df
```

```python
df['wicket_kind'] = df['wicket_kind'].fillna('No Wicket')
df['player_out'] = df['player_out'].fillna('None')
df['fielders'] = df['fielders'].fillna('None')

df['extra_type'] = df['extra_type'].fillna('None')

df['review_decision'] = df['review_decision'].fillna('No Review')
df['umpire'] = df['umpire'].fillna('Unknown')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
numeric_cols = [
    'runs_batter', 'runs_total', 'runs_extras',
    'team_runs', 'team_wicket'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad'
}

df['batting_team'] = df['batting_team'].replace(team_mapping)
df['bowling_team'] = df['bowling_team'].replace(team_mapping)
df['toss_winner'] = df['toss_winner'].replace(team_mapping)
df['match_won_by'] = df['match_won_by'].replace(team_mapping)
```

```python
# Ball number in innings
df['total_balls'] = df['over'] * 6 + df['ball']

# Is boundary
df['is_boundary'] = df['runs_batter'].apply(lambda x: 1 if x in [4,6] else 0)

# Is wicket
df['is_wicket'] = df['wicket_kind'].apply(lambda x: 0 if x == 'No Wicket' else 1)

# Strike rate per ball (approx feature)
df['strike_rate'] = (df['runs_batter'] / df['balls_faced'].replace(0, np.nan)) * 100
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
df.info()
```

```python
df.to_csv('../data/processed/ipl_cleaned.csv', index=False)
```
